```python

x = titanic['fare']
y = titanic['age']

# Create JointGrid
g = sns.JointGrid(data=titanic, y=y, x=x)

# Plot the data
g.plot_joint(sns.scatterplot)
g.plot_marginals(sns.histplot)

# Set x-axis range and ticks
g.ax_marg_x.set_xlim(-100, 600)
g.ax_marg_x.set_xticks(np.arange(-100, 601, 100))

# Set y-axis range and ticks
g.ax_marg_y.set_ylim(-10, 90)
g.ax_marg_y.set_yticks(np.arange(0, 81, 20))

# Show plot
plt.show()
```

```python
# Create swarmplot
plot = sns.swarmplot(data=titanic, x="class", y="age", size=5, hue='class', palette='Set2',legend=False)

# Set y-axis range and ticks
plot.set_ylim(-10, 90)
plot.set_yticks(range(0, 81, 20))

# Remove the hue
plot.set_xticklabels([])


# Add grid lines to y-axis
plot.yaxis.grid(True)


# Show plot
plt.show()
```

```python

# Create figure and axes
fig, ax = plt.subplots(figsize=(5, 5))

# Plot histogram
sns.histplot(data=titanic, x='sex',hue='sex', alpha=1)

# Set x-axis tick labels and positions
ax.set_xticks([0.2, 0.9])


# Set y-axis limits and tick positions
ax.set_ylim(0, 600)
ax.set_yticks(range(0, 601, 100))

# Show y-axis grid lines
ax.yaxis.grid(True)

# Show plot
plt.show()
```
